```
import pandas as rp
import matplotlib.pyplot as mt
import re
```

```
data=rp.read_csv('a.csv')
```

```
bad=data[data.rate=='bad']
```

```
booleans = []
for result in data.rate:
    if  re.search('bad',result):
        booleans.append(True)
supcount=len(booleans)
print(supcount)
```

```
sup=data[data.rate=='super']
```

```
nice=data[data.rate=='nice']
```

```

mt.plot(nice.date,nice.people)
mt.plot(bad.date,bad.people)
mt.plot(sup.date,bad.people)
mt.title("Tech patro Data")
mt.legend(['nice review','bad review','super review'])
mt.ylabel("Numbers of people")
mt.xlabel("date")
mt.show()
```
